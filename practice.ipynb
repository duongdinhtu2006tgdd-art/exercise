### 2. Hiệu ứng......

```python
# PHẦN 1 
import time

print("PART 1".center(30)) 

A = "I love you so much"
print("hi honey:\n", end ="") 

for char in A: 
    print(char, end='', flush=True)  # flush=True: đẩy ký tự ra màn hình ngay lập tức
    time.sleep(0.1) 

print()  # ngắt dòng sau vòng lặp

B = "goodnight my love"
for char in B: 
    print(char, end="", flush=True)
    time.sleep(0.1)
print('\n')


# PHẦN 2
print("PART 2".center(30))
title = 'academic results'

sv1 = "Dương Đình Tú", 10  # Tuple Packing: Gói Tên và Điểm vào 1 biến tuple
sv2 = "Đoàn Minh Tú", 1

print("-" * 30)
print(title.center(30))
print("-" * 30)

print("Name".rjust(10), "Score".rjust(10), sep = "|".rjust(5))
print("-" * 30)

# sv1[0]: tên, str(sv1[1]): ép int -> str để dùng .rjust()
print(sv1[0], str(sv1[1]).rjust(8), sep ="|".rjust(2))
print(sv2[0], str(sv2[1]).rjust(7), sep ="|".rjust(3))
print()


# PHẦN 3: Ghi log ra file
print("PART 3".center(30))

# with open: tự động đóng file sau khi ghi xong
with open("log.txt", 'w', encoding='utf-8') as f:  # encoding='utf-8': tránh lỗi font tiếng Việt
    print("LOG Hệ Thống", file=f)
    print("Thời gian chạy:", time.asctime(), file=f)
    print("Trạng thái: Hoàn thành", file=f)

C = "đã ghi thông tin vào file log.txt thành công!"
for char in C:
    print(char, end="", flush=True)
    time.sleep(0.1)
print()
```

### 3. Đọc cảm biến

```python
# dữ liệu đầu vào - giả lập độ ẩm đất
do_am_dat = [65, 70, -1, 45, 80, 20, 150, 55]

# xử lý dữ liệu đầu vào
idx = 0
while idx < len(do_am_dat):
    if do_am_dat[idx] < 0:
        print(f"lỗi tín hiệu tại vị trí {idx}") 
        idx += 1
        continue

    elif do_am_dat[idx] > 100: 
        print("CẢNH BÁO: Chập mạch cảm biến! Ngắt hệ thống")
        break
    
    elif 0 <= do_am_dat[idx] <= 100:
        print(f"vị trí {idx}: có độ ẩm đất là {do_am_dat[idx]} %")
        idx += 1

# else chạy nếu while kết thúc tự nhiên (không bị break)
else: 
    print("Hoàn tất chu kỳ đọc cảm biến. Mạch chuyển sang chế độ ngủ")
```
